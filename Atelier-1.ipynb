## Étape 2 - Affichage d'un tableau

```java
System.out.println("+------------+-------+-----+------+");
System.out.println("| Produit    | PU    | Qté | Prix |");
System.out.println("+------------+-------+-----+------+");
System.out.println("| Stylos     |  1.20€|    5| 6.00€|");
System.out.println("| Cahiers    |  3.50€|    2| 7.00€|");
System.out.println("+------------+-------+-----+------+");
```

## Étape 3 - Affichage de variables

```java
String nomProduit1 = "Stylos";
double prixUnitaire1 = 1.20;
int quantite1 = 5;
String nomProduit2 = "Cahiers";
double prixUnitaire2 = 3.50;
int quantite2 = 2;

double total1 = prixUnitaire1 * quantite1;
double total2 = prixUnitaire2 * quantite2;
double total = total1 + total2;

System.out.println("\nTableau généré dynamiquement avec variables :\n");
System.out.println("+------------+--------+------+--------+");
System.out.printf("|  %-10s|  %5.2f€|   %3d|   %4.2f€|\n", nomProduit1, prixUnitaire1, quantite1, total1);
System.out.printf("|  %-10s|  %5.2f€|   %3d|   %4.2f€|\n", nomProduit2, prixUnitaire2, quantite2, total2);
System.out.println("+------------+--------+------+--------+");
System.out.printf("| TOTAL      |        |      |  %.2f€|\n", total);
System.out.println("+------------+--------+------+--------+");
```

## Étape 4 - Saisie des informations

```java
Scanner clavier = new Scanner(System.in);

System.out.print("Entrez la quantité pour " + nomProduit1 + " : ");
quantite1 = clavier.nextInt();
System.out.print("Entrez la quantité pour " + nomProduit2 + " : ");
quantite2 = clavier.nextInt();

total1 = prixUnitaire1 * quantite1;
total2 = prixUnitaire2 * quantite2;
total = total1 + total2;
       
System.out.println("\nTableau mis à jour après saisie utilisateur :\n");
System.out.println("+------------+---------+-----+------------+");
System.out.printf("| %-10s |   %5.2f€|  %3d|   %8.2f€|\n", nomProduit1, prixUnitaire1, quantite1, total1);
System.out.printf("| %-10s |   %5.2f€|  %3d|   %8.2f€|\n", nomProduit2, prixUnitaire2, quantite2, total2);
System.out.println("+------------+---------+-----+------------+");
System.out.printf("| TOTAL      |         |     |   %8.2f€|\n", total);
System.out.println("+------------+---------+-----+------------+");
clavier.close();
```
